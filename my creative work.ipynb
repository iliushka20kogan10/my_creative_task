Здесь простой пример использования функций librosa.

Импортируем librosa.

```python
import librosa
```

Создаём переменную, которая будет хранить в себе путь к аудио.

```python
filename = librosa.example('nutcracker')
```

Загрузим аудио как сигнал "у". Сохраним частоту дискретизации в "sr".

```python
y, sr = librosa.load(filename)
```

Запустим бит-трекер. Выходные данные трекера ритма представляют собой оценку темпа (в ударах в минуту) и массив номеров кадров, соответствующих обнаруженным событиям ритма.

```python
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
```

Следующая операция преобразует номера кадров beat_framesв тайминги:

```python
beat_times = librosa.frames_to_time(beat_frames, sr=sr)
```

Теперь beat_times будет массив временных меток (в секундах), соответствующих обнаруженным событиям биения.
Можно вывести beat_times, чтобы посмотреть:

```python
print(beat_times)
```
